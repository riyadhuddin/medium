```python
# -*- coding: utf-8 -*-
"""
Created on Mon Dec 23 11:29:06 2019
@author: Riyadh Uddin
"""
from collections import Counter
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as seabornInstance 

paragraph = """I was hoping to start writing something nice for long time though I could not decide what to write about or you can say I don't get any nice creative thought to assign for my wild dream. Yet started to write it in the name of stupidity, Lets see how long I can hang or quit. You probably wouldn't like to know me in details,so lets just forget it except my name :p Riyadh Uddin.
"""
wordcount = len(paragraph.split()) 
print ("The number of words in paragraph are : " + str(wordcount)) 
number = Counter(paragraph) 
print ("Count of I in paragraph is : "
                       +  str(number['I']))
counts = dict()
words = paragraph.split()
for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
```

```python
datawords = counts
datasetword =pd.DataFrame(list(datawords.items()),columns = ['Words','Frequency']) 
datachar = number
datasetchar = pd.DataFrame(list(datachar.items()),columns = ['Character','Frequency']) 
```

```python
datasetword.plot(x='Words', y='Frequency', style='o')  
plt.title('Words vs Frequency')  
plt.xlabel('Words')  
plt.ylabel('Frequency')  
plt.show()
```

```python
datasetchar.plot(x='Character', y='Frequency', style='o')  
plt.title('Character vs Frequency')  
plt.xlabel('Character')  
plt.ylabel('Frequency')  
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(datasetword['Frequency'])
```
